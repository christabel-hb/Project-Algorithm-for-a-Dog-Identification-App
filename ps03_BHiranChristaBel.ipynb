# **Exercise 1. Basic Word Tokenization**

[link text](https://)Tokenization with NLTK library:
Tokenization is the first step in text processing.

**Step 1:** Install NLTK

```python
!pip install nltk
```

Import NLTK library
and download the 'punkt' resource required for tokenization.

```python
import nltk
nltk.download('punkt')
```

**Step 2 :** Import the word_tokenize function from nltk.tokenize.

```python
from nltk.tokenize import word_tokenize
```

 **Step 3 :** Define text in a variable - "Conversational AI is a type of artificial intelligence (AI) that can simulate human conversation. It is made possible by natural language processing (NLP), a field of AI that allows computers to understand and process human language and Google's foundation models that power new generative AI capabilities. It’s constantly learning from its interactions and improving its response quality over time."

```python
text="""Conversational AI is a type of artificial intelligence (AI) that can simulate human conversation. It is made possible by natural language processing (NLP), a field of AI that allows computers to understand and process human language and Google's foundation models that power new generative AI capabilities. It’s constantly learning from its interactions and improving its response quality over time."""
```

**Step 4 :** Use the word_tokenize function to tokenize the text into words

```python
tokens = word_tokenize(text)
```

**Step 5** : Print the list of tokens.

```python
print("Tokens:", tokens)
```

**Step 6 :** Count the number of tokens in the text.

```python
token_count = len(tokens)
print("Number of tokens:", token_count)
```

**Step 7 :** Identify the frequency of each token using nltk.FreqDist or other function.

```python
from nltk.probability import FreqDist
freq_dist = FreqDist(tokens)
print("Token frequencies:", freq_dist)
```

**EXTRA :** Print frequencies in a more readable format

```python
for token, frequency in freq_dist.items():
    print(f"{token}: {frequency}")
```

# Mask :

Researching how to use the mask parameter to create a word cloud in a specific shape.
And uploading a mask image to Drive and using it in  code.

```python
#Load the Mask Image from Google Drive

mask = imageio.imread("/content/drive/My Drive/ConvoAI/star.jpg")

#Load Your Text File
with open("/content/drive/My Drive/ConvoAI/JaneEyre.txt", "r") as file:
    text = file.read()


# Create the Word Cloud with Mask
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",  # Try changing this, e.g., "black"
    colormap="inferno",
    stopwords=set(STOPWORDS),
    mask=mask,              # Use the loaded mask image
    contour_color='pink',
    contour_width=1
).generate(text)

#Display the Word Cloud
plt.figure(figsize=(8, 8))  # Set figure size
plt.imshow(wordcloud, interpolation="bilinear")  # Display the word cloud image
plt.axis("off")  # Turn off the axes
plt.tight_layout(pad=0)  # Adjust layout
plt.show()  # Show the plot
```
